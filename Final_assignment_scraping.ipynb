```python
from bs4 import BeautifulSoup
import pandas as pd
import urllib.request as url
import pandas as pd
#getting data from website (www.yelp.ca)
source = url.urlopen('https://www.yelp.ca/biz/pray-tell-toronto?osq=Bars')
#parsing data using beautifulsoup
page_soup = BeautifulSoup(source, 'html.parser')
#store processed data
reviewdata = []
#getting desired data
mains = page_soup.find_all("div", {"class": "lemon--div__373c0__1mboc arrange-unit__373c0__1piwO arrange-unit-grid-column--8__373c0__2yTAx border-color--default__373c0__2oFDT"})
#looping data for seperating required feilds and add to storage
for main in mains:
    review = {}

    try:
        ratings = main.find("span", {"class": "lemon--span__373c0__3997G display--inline__373c0__3JgLR border-color--default__373c0__MD4Lj"}).div.get('aria-label')
        rating=ratings[:1]
        review['Rating'] = rating
        
    except:
       
        print("no rating")
    try:
        reviews = main.find("p", {"class": "lemon--p__373c0__3Qnnj text__373c0__2pB8f comment__373c0__3EKjH text-color--normal__373c0__K_MKN text-align--left__373c0__2pnx_"}).text
        
        review['Comments'] = reviews
    except:
        print("no review")
    try:
        image =main.find("img", {"class": "lemon--img__373c0__3GQUb photo-box-img__373c0__12uK5"}).get("src")
        
        review['image'] = image
    except:
        print("No image") 
    print("review==================")
    
    reviewdata.append(review)
    print(reviewdata)

#Create dataframe and store data into it
df = pd.DataFrame(reviewdata,columns=['Rating','Comments','image'])
df.head(20)
#push data into csv file
df.to_csv('/Users/shanmukhkancharla/Desktop/Study/Sem 2/1007 Social Data Mining/Final Assignment/bardata.csv',index=False, encoding = 'utf-8')
```
